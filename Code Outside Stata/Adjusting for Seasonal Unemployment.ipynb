```r
nwhtun <- "nwht_unemployment_nonadjusted"
raw <- read_csv(nwhtun, show_col_types = FALSE)
# Harmonize to a monthly ts object with frequency = 12 -------------------------
# Adjust this block depending on your file structure
if (all(c("Year", "Month", "Unemployment Rate") %in% names(raw))) {
  raw <- raw %>% mutate(date = as.Date(sprintf("%04d-%02d-01", Year, Month))) %>% arrange(date)
  start_year <- min(raw$Year)
  start_month <- min(raw$Month_num[raw$Year == start_year])
  ts_data <- ts(raw$Unemployment_Rate, start = c(start_year, start_month), frequency = 12)
} else {
  stop("CSV must contain either (Year,Month,Unemployment Rate) columns.")
}
```
