```python
import pandas as pd

# Importing Black, White, and Nonwhite Unemployment
whtun1_df = pd.read_csv("SeriesReport-20250815154743_e7c08a.csv")
nwhtun1_df = pd.read_csv("SeriesReport-20250815133656_d63a3f.csv")
blkun1_df = pd.read_csv("SeriesReport-20250815132626_65ff1a.csv")


# Drop rows of introduction
whtun2_df = whtun1_df.drop(range(11))
nwhtun2_df = nwhtun1_df.drop(range(15))
blkun2_df = blkun1_df.drop(range(11))

# Use the first row as column headers
whtun2_df.columns = whtun2_df.iloc[0]      # set headers
whtun3_df = whtun2_df.drop(11).reset_index(drop=True)  # drop the header row
nwhtun2_df.columns = nwhtun2_df.iloc[0]      # set headers
nwhtun3_df = nwhtun2_df.drop(15).reset_index(drop=True)  # drop the header row
blkun2_df.columns = blkun2_df.iloc[0]      # set headers
blkun3_df = blkun2_df.drop(11).reset_index(drop=True)  # drop the header row

# Getting quartely data
month_to_quarter = {
    "Jan": "Q1", "Feb": "Q1", "Mar": "Q1",
    "Apr": "Q2", "May": "Q2", "Jun": "Q2",
    "Jul": "Q3", "Aug": "Q3", "Sep": "Q3",
    "Oct": "Q4", "Nov": "Q4", "Dec": "Q4"
}
#Changing data format
whtun4_df = whtun3_df.melt(
    id_vars="Year",
    var_name="Month",
    value_name="Unemployment_Rate"
)
whtun4_df["Quarter"] = whtun4_df["Month"].map(month_to_quarter)

nwhtun4_df = nwhtun3_df.melt(
    id_vars="Year",
    var_name="Month",
    value_name="Unemployment_Rate"
)
nwhtun4_df["Quarter"] = nwhtun4_df["Month"].map(month_to_quarter)

blkun4_df = blkun3_df.melt(
    id_vars="Year",
    var_name="Month",
    value_name="Unemployment_Rate"
)
blkun4_df["Quarter"] = blkun4_df["Month"].map(month_to_quarter)

import pandas as pd

# month order mapping
month_order = {
    "Jan": 1, "Feb": 2, "Mar": 3,
    "Apr": 4, "May": 5, "Jun": 6,
    "Jul": 7, "Aug": 8, "Sep": 9,
    "Oct": 10, "Nov": 11, "Dec": 12
}

# add a helper column for sorting
whtun4_df["Month_num"] = whtun4_df["Month"].map(month_order)
nwhtun4_df["Month_num"] = nwhtun4_df["Month"].map(month_order)
blkun4_df["Month_num"] = blkun4_df["Month"].map(month_order)

# sort by Year then Month_num
whtun4_df = whtun4_df.sort_values(["Year", "Month_num"]).reset_index(drop=True)
nwhtun4_df = nwhtun4_df.sort_values(["Year", "Month_num"]).reset_index(drop=True)
blkun4_df = blkun4_df.sort_values(["Year", "Month_num"]).reset_index(drop=True)




whtun4_df.to_csv("wht_unemployment_adjusted.csv", index=False)
nwhtun4_df.to_csv("nwht_unemployment_nonadjusted.csv", index=False)
blkun4_df.to_csv("blk_unemployment_adjusted.csv", index=False)
```
